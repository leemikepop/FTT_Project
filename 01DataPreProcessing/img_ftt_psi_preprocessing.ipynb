Create a folder with training images

```python
df = pd.read_csv("./ftt_psi_original.csv")
for i in range(1,df.shape[0],2):
    df.iloc[i,0] = df.iloc[i,0]+"-01"
    df.iloc[i+1,0] = df.iloc[i+1,0]+"-02"
name_list = df.iloc[:,0].to_list()
# df.to_csv("./ftt_psi_original.csv")
```

```python
dst = "./train_img_%.0f/"%time.time()
# dst = "./train_img_out_%.0f"%time.time()
os.makedirs(dst)
TYPE = ["WO1","CW2","WK3","LF4"]
filenames = []
i = 0
for type in TYPE:
    src_path = "./%s_copy/"%type
    # src_path = "./%s_copy_out"
    pathlist = Path(src_path).glob(r'*.jpg')
    for path in pathlist:
        name = os.path.basename(path)
        if name[:-4] in name_list:
            filenames.append(name[:-4])
            print(filenames[-1])
            # shutil.copy(path,"%s%03d_%s"%(dst,i+1,name))
            i = i + 1
```

Read original csv file & rename

```python
drop_list = []
for i in range(1,len(name_list),1):
    if name_list[i] not in filenames:
        drop_list.append(i)
df = df.drop(index=drop_list)
df.to_csv("./ftt_psi.csv",index=False,header=False)
```
